```python
import csv
import matplotlib.pyplot as plt

f = open("city_ev.csv")
data = csv.reader(f)
next(data)

city = ['', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '']
city_ev = [0] * 17

for row in data:
    if '' in row[4]:
        for i in range(len(city)):
            if row[i+5] != '  ':
                city_ev[i] += int(row[i+5])
    
#print(city)
#print(city_ev)


f = open("charger.csv")
data = csv.reader(f)
next(data)

cg_city = ['', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '']
cg_count = [0] * 17

for row in data:
    for i in range(len(cg_city)):
        if row[2].split(' ')[0] in cg_city[i]:
            if row[3] != '':
                cg_count[i] += int(row[3])
    
#print(cg_city)
#print(cg_count)

plt.rc('font', family='Malgun Gothic')
plt.figure(dpi=300)
plt.barh(city, city_ev, label = '/  ',color='#d4eae8')
plt.barh(city, cg_count, label = '/  ',color='#18caca')
plt.title(' /     ')
plt.xlabel('',labelpad=10)
plt.ylabel('',labelpad=10)
plt.legend()
plt.show()
```
